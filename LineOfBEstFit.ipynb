```python
import pandas as pd

FinalSuperset = pd.read_csv('FinalSuperset\FinalSuperset_joined.csv')
FinalSuperset_N = pd.read_csv('FinalSuperset\FinalSuperset_joined_Nminmax.csv')
target = "API"

mean_altitude_2017 = FinalSuperset[FinalSuperset['Year'] == 2017][target].mean()
mean_altitude_2018 = FinalSuperset[FinalSuperset['Year'] == 2018][target].mean()
mean_altitude_2019 = FinalSuperset[FinalSuperset['Year'] == 2019][target].mean()
mean_target_2017_N = FinalSuperset_N[FinalSuperset_N['Year'] == 2017][target].mean()
mean_target_2018_N = FinalSuperset_N[FinalSuperset_N['Year'] == 2018][target].mean()
mean_target_2019_N = FinalSuperset_N[FinalSuperset_N['Year'] == 2019][target].mean()

print(mean_altitude_2017)
print(mean_altitude_2018)
print(mean_altitude_2019)

# Line of Best Fit
import numpy as np

# Create a DataFrame
df = pd.DataFrame({
    'X': [2017, 2018, 2019],
    'Y': [mean_altitude_2017, mean_altitude_2018, mean_altitude_2019]
})


# Assuming df is your DataFrame and 'X' and 'Y' are your columns
X = df['X'].values
Y = df['Y'].values

# Perform linear regression (1st degree polynomial)
coefficients = np.polyfit(X, Y, 1)

# The first element in the returned array is the slope (gradient)
gradient = coefficients[0]

print("The gradient of the line of best fit is: ", gradient)

# Plot the line of best fit
import matplotlib.pyplot as plt

plt.title("Normalized")
plt.scatter(X, Y)
plt.plot(X, gradient*X + coefficients[1], color='red')
plt.show()


df2 = pd.DataFrame({
    'X': [2017, 2018, 2019],
    'Y': [mean_target_2017_N, mean_target_2018_N, mean_target_2019_N]
})

X = df2['X'].values
Y = df2['Y'].values

coefficients = np.polyfit(X, Y, 1)

gradient = coefficients[0]

print("The gradient of the line of best fit is: ", gradient)

plt.scatter(X, Y)
plt.plot(X, gradient*X + coefficients[1], color='red')
plt.show()
```
